```python
power_df = pd.read_csv(f"{ROOT}/data/raw_meters/Norwood_25_Ton_RTU_9_480V100A.csv", header=1)
power_df.head()
```

```python
temps_df = pd.read_csv(f"{ROOT}/data/raw_meters/Norwood_Tool_Co_RTU_9_Temps.csv", header=1)
temps_df.head()
```

```python
temps_df.columns
```

```python
column_name_map = {
    'Date Time, GMT-04:00': "Datetime",
    'Temp, °F (LGR S/N: 10626338, SEN S/N: 10626338, LBL: OAT)': "OAT",
    'Temp, °F (LGR S/N: 10626338, SEN S/N: 10626338, LBL: Mixed T)': "Mixed Temp",
    'Temp, °F (LGR S/N: 10626338, SEN S/N: 10626338, LBL: Return T)': "Return Temp",
    'Temp, °F (LGR S/N: 10626338, SEN S/N: 10626338, LBL: Supply T)': "Supply Temp",
}
temps_df.rename(
    columns=column_name_map, inplace=True
)
temps_df = temps_df[[col_name for col_name in column_name_map.values()]]
temps_df["Datetime"] = pd.to_datetime(temps_df["Datetime"])
temps_df.set_index("Datetime", inplace=True)
temps_df.head()
```

```python
power_df.columns
```

```python
column_name_map = {
    'Date Time, GMT-04:00': "Datetime",
    'Counts, # (LGR S/N: 10237654, SEN S/N: 10237654, LBL: Norwood 25 Ton RTU 9 480V100A)': "Counts",
#     'Energy, kWh (LGR S/N: 10237654)': "kWh", # kWh will get messy, better to use avg kW
    'Avg Power, kW (LGR S/N: 10237654)': "kW",
}
power_df.rename(
    columns=column_name_map, inplace=True
)
power_df = power_df[[col_name for col_name in column_name_map.values()]]
power_df["Datetime"] = pd.to_datetime(power_df["Datetime"])
power_df.set_index("Datetime", inplace=True)
power_df.head()
```

```python
temps_df_hourly = temps_df.resample("H").mean()
temps_df_hourly.head()
```

```python
power_df_hourly = power_df.resample("H").mean()
power_df_hourly.head()
```

```python
final_df = power_df_hourly.merge(temps_df_hourly, left_index=True, right_index=True)
final_df.head()
```

```python
final_df.shape
```

```python
final_df.to_csv(f"{ROOT}/data/power_and_oat.csv")
```
